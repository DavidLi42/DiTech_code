```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display # Allows the use of display() for DataFrames

# Show matplotlib plots inline (nicely formatted in the notebook)
%matplotlib inline

import time
import datetime
import os

#import datacleaningp1 as dcp
import dcp
reload(dcp)

data_train_all = pd.read_csv('TrainingData_fixed_4_test_1.csv')
data_test_all = pd.read_csv('test_data_1.csv')

def aligncols(data,col_list):
    output = pd.DataFrame(index = data.index)
    for col in col_list:
        col_data = data[col]
        output = output.join(col_data)
        
    return output

col_standard = ['district_id','time','gaps', 'gaps_pre_1', 'gaps_pre_2', 'gaps_pre_3',\
                'PM2.5', 'Weather_1', 'Weather_2', 'Weather_3', \
                'Weather_4', 'Weather_6', 'Weather_8', 'Weather_9','temperature', \
                'tj_level_1', 'tj_level_2', 'tj_level_3', 'tj_level_4']

def output_required_format(data):
    '''rewirte data into required format'''
    output = pd.DataFrame(index = data.index)
    #row_number = len(data)
    ### add district_id
    output['district_id'] = data['district_id']
    ### add date-time
    date_time_list = []
    for row_id in data.index:
        date_tmp = data.loc[row_id]['date']
        time_tmp = str(int(data.loc[row_id]['time']))
        date_time_list.append(date_tmp + '-' + time_tmp)
        
    output['date-time'] = date_time_list
        
    ### add gaps_pred
    output['gaps_pred'] = data['gaps']
    ### reset index
    output = output.reset_index(drop = True)
    return output
```

```python
district_number = 66
#test_set_size = 0.3 #use all training data
num_neighbors = 10

cols_output = ['district_id','date-time','gaps_pred']
output_test_pred = pd.DataFrame(columns = cols_output)

#data_train_all = data_train_all_v2

### import necessary moduels in sklearn
from sklearn.utils import shuffle
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,r2_score
###

for iid in np.arange(district_number)+1:
    ### select training data for one specific district
    data_train_iid = data_train_all[data_train_all['district_id'] == iid]
    data_train_iid.drop(['district_id'],inplace = True, axis = 1)
    data_train_iid = data_train_iid.reset_index(drop = True)
    
    ### select test data for one specific district
    data_test_iid = data_test_all[data_test_all['district_id'] == iid]
    data_test_iid_backup = data_test_iid.copy()
    #data_test_iid.drop(['district_id'],inplace = True, axis = 1)
    data_test_iid = data_test_iid.reset_index(drop = True)
    
    ### store date information in another DataFrame
    data_test_iid_date = data_test_iid['date']
    
    ### drop dateinfo and dummy'gaps' from data_test_iid, align its columns according to col_standard
    data_test_iid.drop(['date'], inplace = True, axis = 1)
    data_test_iid = aligncols(data_test_iid,col_standard)
    data_test_iid.drop(['gaps','district_id'],inplace = True, axis = 1)
    
    
    ### shuffle training data, prepare to apply nn model
    data_train_iid_s = shuffle(data_train_iid,random_state = 42)
    ### split labels from data set
    label = data_train_iid_s['gaps']
    data_train_iid_s.drop(['gaps'],axis = 1,inplace = True)
    
    ### feature renormalization on training_set,then on test_set
    min_max_scaler = preprocessing.MinMaxScaler()
    data_train_iid_n = min_max_scaler.fit_transform(data_train_iid_s)
    data_test_iid_n = min_max_scaler.transform(data_test_iid)

    ### train a nearest neighbor regressor
    neigh = KNeighborsRegressor(n_neighbors= num_neighbors)
    neigh.fit(data_train_iid_n,label)
    
    y_pred = neigh.predict(data_test_iid_n)
    data_test_iid_backup['gaps'] = y_pred
    
    ### transform output into required format
    output_tmp = output_required_format(data_test_iid_backup)
    output_test_pred = pd.concat([output_test_pred,output_tmp],ignore_index = True)
    #print 'pred test set %d done '%iid
    #print '\n'

test_real = pd.read_table('test_result_1',sep = ',',names = ['district_id','date-time','gaps_real'])
test_result_all = pd.merge(test_real,output_test_pred,how = 'left',on = ['district_id','date-time'])
#display(test_result_all)
print mean_absolute_error(test_result_all['gaps_real'],test_result_all['gaps_pred'])

output_test_pred['district_id'] = output_test_pred['district_id'].apply(dcp.num2str)
#display(output_test_pred)
output_test_pred.to_csv('test_pred_1.csv',index = False)
```
